```python
# If you want to choose the color you need to search in google for html code and give it with #code.
# Aplha is to change the intensity of the color.
# marker to represent the dots
plt.scatter(x,y, c = '#1154EC', alpha = .9, marker = 'x')
plt.xlabel("This is my x- axis")
plt.ylabel("This is my y- axis")
plt.title("My graph")
plt.grid() # to represent with grid.
plt.figure(figsize = (6,2)) # To change the size of the graph.
```

```python
x = ['a','b','s','d','e']
y= np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize= (5,3))
plt.bar(x,y, color = 'b')
plt.xlabel("categorical data")
plt.ylabel("numerical data")
plt.title("my data")
plt.grid()
```

```python
plt.figure(figsize= (5,3))
plt.barh(x,y, color = 'b') # barh changes the whole data into horizontal.
plt.xlabel("Categorical data")
plt.ylabel("Numerical data")
plt.title("My data")
plt.grid()
```

```python
data = [1,2,3,4,5,6,7,8,9,4,5,6,2,1,8,8,8,8,6,6,7,7,7,5]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data, color= "b") # in the below histogram we can understand that on x axis we have distinct data and on y axis we can see how many times it has been repeated
plt.xlabel("representing unique values")
plt.ylabel("representing freq")
plt.savefig("test.png")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
# 3d plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c = 'r', alpha = .5)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot(x,y,z)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
